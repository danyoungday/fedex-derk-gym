```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
gens = ["0", "2", "5", "10", "14", "19"]
colors = ["#FF0000", "#FF7F00", "#FFFF00", "#00FF00", "#0000FF", "#9400D3"]
values = [4.48, 8.68, 2.40, 12.53, 2.77, 0.87]

plt.scatter(gens, values, c=colors)
plt.plot(gens, values, color="gray")
plt.xlabel("Generation")
plt.ylabel("Avg Reward Across 10 Fights")
plt.show()
```

```python
df = pd.read_csv("results/log.csv")
```

```python
df.mean(axis=1).plot(marker="o", label="mean")
df.max(axis=1).plot(marker="o", label="max")
plt.xlabel("Generation")
plt.xticks(np.arange(0, 20, 1))
plt.ylabel("Average Reward Across 4 Fights")
plt.grid()
plt.legend()
plt.show()
```
